```python
from pythreader import PyThread, Escrow, Primitive
import time
from threading import Thread
```

```python
escrow = Escrow()

sema = Primitive()
counter = 1


def get():
    global sema
    item = escrow.get()
    with sema:
        print("    get:", item)
    
def put():
    global sema, counter
    with sema:
        t = counter
        counter += 1
        
    escrow.put(t)
    with sema:
        print("put:", t)

    
```

```python
readers = [Thread(target=get) for _ in range(100)]
writers = [Thread(target=put) for _ in range(100)]
[r.start() for r in readers]
[w.start() for w in writers]
[t.join() for t in readers+writers]
```
